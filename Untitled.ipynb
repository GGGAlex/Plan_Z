```python
import pandas as pd
import numpy as np
from flask import Flask
from flask import request
from flask_restplus import Resource, Api
from flask_restplus import fields
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
file = pd.read_csv("mushrooms.csv")
df = pd.DataFrame(data=file)
list(df)
```

```python
df_na = df.replace("?", np.nan)
narows = df_na[df_na.isnull().any(axis=1)]
nacols = df_na.columns[df_na.isna().any()].tolist()

# Calculate the percentage of rows and columns that contain a NaN value
percent_narows = round(len(narows)/df_na.shape[0]*100, 1)
percent_nacols = round(len(nacols)/df_na.shape[1]*100, 1)

print("There are {} row(s) ({}%) with NaN values.".format(len(narows), percent_narows))
print("There are {} column(s) ({}%) with NaN values.".format(len(nacols), percent_nacols))
```

```python
cleaned_df = df_na.dropna()
print(len(cleaned_df))
```

```python
narows = cleaned_df[cleaned_df.isnull().any(axis=1)]
nacols = cleaned_df.columns[cleaned_df.isna().any()].tolist()

# Calculate the percentage of rows and columns that contain a NaN value
percent_narows = round(len(narows)/cleaned_df.shape[0]*100, 1)
percent_nacols = round(len(nacols)/cleaned_df.shape[1]*100, 1)

print("There are {} row(s) ({}%) with NaN values.".format(len(narows), percent_narows))
print("There are {} column(s) ({}%) with NaN values.".format(len(nacols), percent_nacols))
```

```python
class_counts = cleaned_df["class"].value_counts()
class_counts.index = ["e", "p"]

print("Percentage of each class:")
print(class_counts/cleaned_df.shape[0] * 100)
```
